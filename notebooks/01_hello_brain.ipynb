# Hello Brain: Visualizing MRI/fMRI Data

Welcome to our first brain visualization notebook. Today we will use Python to look at a standard brain image and explore it slice by slice.

## What is MRI? What is fMRI?

**MRI** stands for **Magnetic Resonance Imaging**. It is a way to make detailed pictures of the inside of the body without using a camera or cutting anything open.

An MRI scanner uses a strong magnet and radio waves. A computer turns the signals into images, like slices through the brain.

**fMRI** stands for **functional Magnetic Resonance Imaging**. It is related to MRI, but it is used to study brain activity. fMRI does **not** directly see thoughts. Instead, it measures changes linked to blood oxygen. These changes can give scientists clues about which brain areas are more active during a task.

## 2. Load a Standard Brain Template

Scientists often compare brain images by placing them into a shared coordinate system. Here we load a standard brain template called **MNI152**.

Think of it like a map: many people can point to the same coordinate and talk about the same location.

```python
# Load a standard brain template included through nilearn.
# This avoids downloading a large dataset.
mni_template = load_mni152_template()

# Print basic information about the image.
print(type(mni_template))
print("Image shape:", mni_template.shape)
```

## 3. Visualize the Brain Template

The next cell shows the brain from three directions at once. The crosshair marks the selected coordinate.

```python
# Plot the anatomical brain template.
plotting.plot_anat(
    mni_template,
    title="MNI152 standard brain template",
    display_mode="ortho",
    cut_coords=(0, 0, 0),
    draw_cross=True,
)

plotting.show()
```

## 4. Voxels: 3D Pixels

A normal digital picture is made of tiny squares called **pixels**.

A brain image is 3D, so it is made of tiny cubes called **voxels**. You can think of voxels as **3D pixels**.

Each voxel stores a number. In this anatomical template, the number helps describe how bright that tiny part of the brain image should be.

```python
# Convert the brain image into a NumPy array of numbers.
brain_data = mni_template.get_fdata()

# Look at the full 3D shape and one example voxel value.
print("3D image shape:", brain_data.shape)
print("Example voxel value:", brain_data[90, 108, 90])
```

## 5. Try Different Brain Slices

`cut_coords` controls where we slice through the brain. The three numbers mean:

- **x**: left to right
- **y**: back to front
- **z**: bottom to top

```python
# Try changing these numbers.
x = 0
y = -30
z = 20

plotting.plot_anat(
    mni_template,
    title=f"Brain slices at x={x}, y={y}, z={z}",
    display_mode="ortho",
    cut_coords=(x, y, z),
    draw_cross=True,
)

plotting.show()
```

```python
# Compare several front-to-back slices.
for y in [-60, -30, 0, 30, 60]:
    plotting.plot_anat(
        mni_template,
        title=f"Front/back coordinate y={y}",
        display_mode="y",
        cut_coords=[y],
    )
    plotting.show()
```

## 6. Interactive Brain Viewer

Use the sliders to move the slice location. This works in Colab and in many local Jupyter notebooks.

```python
from ipywidgets import IntSlider, interact


def show_brain(x=0, y=0, z=0):
    """Show the brain at one x, y, z coordinate."""
    plotting.plot_anat(
        mni_template,
        title=f"x={x}, y={y}, z={z}",
        display_mode="ortho",
        cut_coords=(x, y, z),
        draw_cross=True,
    )
    plotting.show()


interact(
    show_brain,
    x=IntSlider(value=0, min=-70, max=70, step=5, description="x"),
    y=IntSlider(value=0, min=-100, max=80, step=5, description="y"),
    z=IntSlider(value=0, min=-50, max=90, step=5, description="z"),
);
```
