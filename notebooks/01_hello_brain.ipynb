## What is MRI? What is fMRI?

**MRI** stands for **Magnetic Resonance Imaging**. It is a way to make detailed pictures of the inside of the body without using a camera or cutting anything open.

An MRI scanner uses a strong magnet and radio waves. A computer turns the signals into images, like slices through the brain.

**fMRI** stands for **functional Magnetic Resonance Imaging**. It is related to MRI, but it is used to study brain activity. fMRI does **not** directly see thoughts. Instead, it measures changes linked to blood oxygen. These changes can give scientists clues about which brain areas are more active during a task.

## Voxels: 3D Pixels

A normal digital picture is made of tiny squares called **pixels**.

A brain image is 3D, so it is made of tiny cubes called **voxels**. You can think of voxels as **3D pixels**.

Each voxel stores a number. In this anatomical template, the number helps describe how bright that tiny part of the brain image should be.

## Brain Slices and Directions

A brain image is like a stack of many thin slices. We can look through the stack from different directions.

Scientists often describe three main slice directions:

- **Sagittal** slices divide left and right.
- **Coronal** slices divide front and back.
- **Axial** slices divide top and bottom.

The interactive viewer below shows these directions together. When you click in one view, the crosshair moves to that point, and the other views jump to the matching slice.

## Interactive Brain Viewer

The viewer below lets you move through the brain with your cursor.

Try clicking inside the brain image. The crosshair moves to the place you clicked, and the other views update too.

The coordinates work a bit like a map inside the brain:

- **x** moves left to right.
- **y** moves back to front.
- **z** moves bottom to top.

When you click, you are choosing one voxel location. The viewer then shows the slices that pass through that location.

```python
# Create one clickable brain viewer.
# Click inside the viewer to move through brain slices.
brain_viewer = plotting.view_img(
    mni_template,
    bg_img=False,
    title="Click to explore the MNI152 brain template",
    symmetric_cmap=False,
    cmap="gray",
    colorbar=False,
)

brain_viewer
```
